```python
#5...
from scipy import stats
import numpy as np

# Given data
mean = 75
std_dev = 12
sample_size = 30  # assuming sample size is 30 for this example
confidence_level = 0.90

# Calculate z-score for the given confidence level
z_score = stats.norm.ppf((1 + confidence_level) / 2)

# Calculate margin of error
margin_of_error = z_score * (std_dev / np.sqrt(sample_size))

# Calculate confidence interval
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

# Output results
print(f"90% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
#6...
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Parameters
df = 10  # degrees of freedom
x = np.linspace(0, 30, 1000)  # range of values for chi-square

# Calculate chi-square PDF
pdf = chi2.pdf(x, df)

# Set the chi-square statistic for shading
chi2_statistic = 15

# Plot the chi-square distribution
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, 'b-', label=f'Chi-square distribution (df={df})')
plt.fill_between(x, 0, pdf, where=(x >= chi2_statistic), color='red', alpha=0.5, label=f'Area for chi-square ≥ {chi2_statistic}')

# Labeling
plt.title('Chi-square Distribution with 10 Degrees of Freedom')
plt.xlabel('Chi-square Statistic')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#7...
from scipy import stats
import numpy as np

# Given data
n = 1000  # sample size
x = 520   # number of people who prefer Coke
confidence_level = 0.99

# Sample proportion
p_hat = x / n

# Z-score for 99% confidence level
z_score = stats.norm.ppf((1 + confidence_level) / 2)

# Standard error
SE = np.sqrt((p_hat * (1 - p_hat)) / n)

# Margin of error
ME = z_score * SE

# Confidence interval
lower_bound = p_hat - ME
upper_bound = p_hat + ME

print(f"99% Confidence Interval for the proportion of people who prefer Coke: ({lower_bound:.4f}, {upper_bound:.4f})")
```

```python
#8...
from scipy.stats import chi2

# Observed frequencies
observed_tails = 45
observed_heads = 100 - observed_tails

# Expected frequencies for a fair coin
expected_heads = 50
expected_tails = 50

# Chi-square statistic
chi2_statistic = ((observed_heads - expected_heads) ** 2 / expected_heads +
                   (observed_tails - expected_tails) ** 2 / expected_tails)

# Degrees of freedom
df = 1

# Significance level
alpha = 0.05

# Critical value from chi-square distribution
critical_value = chi2.ppf(1 - alpha, df)

print(f"Chi-square Statistic: {chi2_statistic:.4f}")
print(f"Critical Value: {critical_value:.4f}")

# Conclusion
if chi2_statistic > critical_value:
    print("Reject the null hypothesis. There is evidence that the coin is biased.")
else:
    print("Fail to reject the null hypothesis. There is no evidence that the coin is biased.")
```

```python
#9...
from scipy import stats
import numpy as np

# Given data
sample_mean = 72
hypothesized_mean = 70
sample_std_dev = 10
sample_size = 30
alpha = 0.05

# Calculate the t-statistic
t_statistic = (sample_mean - hypothesized_mean) / (sample_std_dev / np.sqrt(sample_size))

# Degrees of freedom
df = sample_size - 1

# Critical t-value for a two-tailed test
critical_value = stats.t.ppf(1 - alpha / 2, df)

print(f"T-statistic: {t_statistic:.4f}")
print(f"Critical value: {critical_value:.4f}")

# Conclusion
if abs(t_statistic) > critical_value:
    print("Reject the null hypothesis. The population mean is significantly different from 70.")
else:
    print("Fail to reject the null hypothesis. There is no significant evidence that the population mean is different from 70.")
```
